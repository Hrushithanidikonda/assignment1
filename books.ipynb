```python
#defining the scatter function
def Books_Data_Clean(data, col_6, col_7):
    '''
    the above function is used to create scatter plot for avg_rating and ratings_count
    
    Returns
    result shows scatter plot between avg_rating and ratings_count
    
    '''
#plotting data to create a scatter plot
    plt.figure(figsize = (10,6))
#plt.scatter(books['avg_rating'], books['ratings_count'], color = 'skyblue') 

    ax = sns.scatterplot(x  = 'avg_rating', y = 'ratings_count', data = books, color = 'skyblue', hue = 'genre')
    plt.xlabel('average rating of books', labelpad = 18, fontweight='bold')
    plt.ylabel('ratings_count', labelpad = 18, fontweight='bold')
    plt.title('Relationship Between Average Ratings and Ratings Count for Books', pad = 20, fontweight='bold')
#calling the function
Books_Data_Clean(books, 'avg_rating', 'ratings_count')
```

```python
#defining the bar function
def Books_Data_Clean(data, col_4, col_11):
    '''
    the above function is used to create bar graph for language code and sales price
    
    Returns
    result shows bar graph between language code and sales price
    
    '''
#plottting data to show the bar graph
    plt.figure(figsize = (8,6))
    sns.barplot(x = 'code', y = 'sale_price', data = books, errorbar = None)

    plt.xlabel('Language code', labelpad = 18,fontweight='bold')
    plt.ylabel('Sales Price', labelpad = 18, fontweight='bold')
    plt.title('Comparison of Sales Price across Language Codes', pad = 20, fontweight='bold')

    plt.tight_layout()
    plt.show()
#calling the function
Books_Data_Clean(books, 'code', 'sale_price')
```

```python
#defining the box plot function
def Books_Data_Clean(data, col_8, col_9):
    '''
    the above function is used to create box plot for genres and gross sales
    
    Returns
    result shows box plot between genres and gross sales
    
    '''
    
#plotting the data for showing the box plot
    plt.figure(figsize = (8,6))
    sns.boxplot(x = 'genre', y = 'gross_sales', data = books, palette = 'YlGnBu' ,showfliers = False) 
            
    plt.xlabel('Genres', labelpad = 18, fontweight='bold')
    plt.ylabel('Gross sales', labelpad = 18, fontweight='bold')
    plt.title('Distribution of Gross Sales Across Different Genres', pad = 20, fontweight='bold')

    plt.tight_layout()
    plt.show()
#calling the function
Books_Data_Clean(books, 'genre', 'gross_sales')
```
